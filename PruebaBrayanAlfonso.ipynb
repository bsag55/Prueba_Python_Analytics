```python
# Identificación estadísticas principales: Media
df2 = df[['SALDO CAPITAL', 'REGIONAL']]
df2 = df2.groupby(['REGIONAL']).mean()
df2
```

```python
# Identificación estadísticas principales: Máx
df2 = df[['SALDO CAPITAL', 'REGIONAL']]
df2 = df2.groupby(['REGIONAL']).max()
df2
```

```python
# Identificación estadísticas principales: Min
df2 = df[['SALDO CAPITAL', 'REGIONAL']]
df2 = df2.groupby(['REGIONAL']).min()
df2.columns
```

```python
# Gráfica Capital por Región

df2 = df[['SALDO CAPITAL', 'REGIONAL']]
df2 = df2.groupby(['REGIONAL']).sum()
df2.plot(kind = 'bar')
```

```python
# Prueba
df2 = df[['SALDO CAPITAL', 'REGIONAL']]
df2['REGIONAL'].value_counts() #Series
df2['REGIONAL'].unique()
df2.groupby('REGIONAL').size() # df
df3 = df[['REGIONAL']]
df3['REGIONAL'].value_counts().tolist()
```

```python
# Número de Registros por región

df4 = df[['SALDO CAPITAL', 'REGIONAL']]
df4 = df4.groupby(['REGIONAL']).size()
df4.plot(kind = 'bar')
```

```python
#  Histograma dias de mora
df_max = df[['DIAS MORA', 'REGIONAL']]
df_max = df.sort_values(by = ['DIAS MORA'], ascending = False)
df_max.head()
df_max = df_max.groupby(['REGIONAL']).head()
#plt.plot()
#df_max.plot(kind = 'bar')
sns.barplot(y ='DIAS MORA',x ='REGIONAL',data=df_max,estimator=max)
```
